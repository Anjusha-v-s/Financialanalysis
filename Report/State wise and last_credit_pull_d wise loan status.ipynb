```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#Importing data to analyse
#In the spreadsheet added a column named 'Year of Last_credit_pull_d' and changed the date format to dd-mm-yyyy
```

```python
data1 = pd.read_csv("Finance_1.csv")
data2 = pd.read_csv("Finance_2.csv")
```

```python
#Merged the two data files
```

```python
Merged = pd.merge(data1, data2, how ='inner' , on = ['id'])
```

```python
#Extracted the required data for analysis
```

```python
Merged[['addr_state','last_credit_pull_d','loan_status']]
```

```python
#state wise loan status
```

```python
statewise = pd.crosstab(Merged['addr_state'],Merged['loan_status'])
statewise
```

```python
#state wise loan status Graphical representation
```

```python
barplot = statewise.plot.bar(rot =0)
```
